# Histogram

A histogram plots given data as per its distribution. It looks quite similar to a bar chart but is different.

Bar charts are best suited for comparing values between different categories or groups, while histograms are used to visualize the distribution of continuous or discrete data.

 A bar chart in Matplotlib is used to compare values between different categories or groups, where each category is represented by a distinct bar. On the other hand, a histogram in Matplotlib is used to visualize the distribution of continuous or discrete data by grouping the data into bins and showing the frequency of data points within each bin. The appearance and purpose of these two types of plots differ based on the type of data and the insights you want to convey.

```python
X = np.random.normal(20,1.5,1000)
```

Now note, in np.random.randn, we are only supposed to pass in either the number of values we want or the dimensions of an array and it generates values according to the standard normal distribution with mean 0 and standard deviation 1.
However, when we want to generate values with our own custom mean and standard deviation, we use np.random.normal(mean, sd, number_of_values) as above

```python
# Histogram

plt.hist(X, color="black", width=1.5)
# the width parameter isn't necessary, by default the histogram will not have spaces between bins
# however, if you lower the width value, it will start looking more like a bar chart
plt.show()
```

Bins are nothing but the different pillars you see in the chart, and they are customizable. For example you can use the parameter '''bins=[X.min(),18,21,X.max()]'''and what this will do is, plot only 3 pillars, one from the min value to 18, one from 18 to 21, and one from 21 to maximum value of x as below.

```python
plt.hist(X, color="black",
         bins=[X.min(),18,21,X.max()])
plt.show()
```

In addition to the custom bins ranges, we can provide the number of bins we want as an argument as well.

```python
plt.hist(X, color="black",
         bins=20)
plt.show()
```

Note we can't count 20 bins to be exact since with a random set of numbers, there are also some negative values which don't show up.

We can also plot a cumulative histogram by using the cumulative parameter set to True.
A cumulative histogram, also known as an "empirical cumulative distribution function" (ECDF), is a graphical representation that shows the cumulative distribution of a dataset. In a cumulative histogram, each point on the graph represents the cumulative count or proportion of data points that have values less than or equal to a particular value on the x-axis.

```python
plt.hist(X, color="black", bins=20,cumulative=True)
plt.show()
```

# Pie Charts

Pie Charts can be used for the following purposes:
*   Proportions within a Whole: Show how parts contribute to a whole, like budget categories.
*   Simple Data Representation: Quick overview for basic information.
*   Percentages and Fractions: Display proportions without numbers.
*   Comparing Categories: Compare relative sizes of a few categories.






```python
lang = ["C++","C#","Python","Java","Go"]
votes = [14,24,50,6,17]
plt.pie(votes, labels=lang)
plt.show()
```

Many a times there can be overlap in data in pie charts due to the presence of too many data categories or categories that share a fairly small percentage. We can apply some transformations through certain parameters to tackle this issue.

```python
explodes =[0,0,0,0.3,0]
plt.pie(votes, labels=lang, explode=explodes, autopct="%.2f%%",pctdistance=1.4)
plt.show()
```

We set up a list explodes to define if we want a particular section to be displayed out of the chart (usually done for small or overlapped categories) and then passed that list as an argument to explode parameter.


Also we set up percentage values for all the categories using autopct(auto percentage) and gave it a string value = "%.2f%%".
The first % tells the autopct that it's not a regular string but a formatted string, .2f means percentage till 2 decimal places in floating point, and %% are used to print % after the percentage value since single % symbol is a strin formatting tool in python.


The pctdistance moves the percentage away from the center of the chart as per given value (done to enhance clarity).

# Box Plots

Box plots in pyplot (Matplotlib) are used to:
*   Detect outliers and unusual data points.
*   Show skewness and symmetry of data.
*   Display central tendency with the median.
*   Represent spread with whiskers and IQR.
*   Provide concise data summarization.
*   Compare distributions of multiple datasets.

They help in quickly understanding a dataset's characteristics, making comparisons, and identifying outliers, all within a single diagram.



```python
# Let us generate 300 heights for this representation
heights = np.random.normal(172,8,300)
```

```python
plt.boxplot(heights)
plt.show()
```

Reading a box plot involves understanding its components and interpreting the information they convey about the distribution of the data. Here's how to read a box plot step by step:
*   Box: The central rectangular box represents the interquartile range (IQR), which contains the middle 50% of the data. The bottom edge of the box is the first quartile (Q1), and the top edge is the third quartile (Q3). The height of the box shows the spread of the middle 50% of the data.
*   Median: Inside the box, a horizontal line represents the median (Q2). The median is the value that separates the lower 50% of the data from the upper 50%.
*   Whiskers: The whiskers extend from the edges of the box to the data points that are within a certain distance from the quartiles. By default, this distance is often set to 1.5 times the IQR. Data points beyond the whiskers are considered potential outliers.

*   Outliers: Individual data points that fall beyond the whiskers are considered potential outliers. They could indicate extreme values or errors in the data.
*   Orientation: The orientation of the box plot can vary. In a vertical box plot, the data values are shown on the y-axis, and the box is oriented vertically. In a horizontal box plot, the data values are shown on the x-axis, and the box is oriented horizontally.
*   Multiple Boxes: If you have multiple box plots side by side, you can compare the distributions of different groups or categories. This is particularly useful for visualizing differences in medians, ranges, and spreads.
*   Skewness: The position of the median within the box can give you an idea of the skewness (asymmetry) of the distribution. If the median is closer to one end of the box, it indicates skewness in that direction.







```python
# Let us understand skewness with an example
first = np.linspace(0,10,25)
second = np.linspace(10,200,25)
third = np.linspace(200,210,25)
fourth = np.linspace(210,230,25)

data = np.concatenate((first,second,third,fourth))

plt.boxplot(data)
plt.show()
```

Now you can see the values are pretty extremely skewed towards 200 since we used data that was kind of biased. We had 25 values from 0 to 10, 25 from 10 to 200 which explains the width of the box, and then we had 50 total values from 200 to 230 that explains the position of the median in the plot.
Thus we can study the characteristics of data.

Suppose you have a dataset where you need to plot different lines in the same chart, you can do that by calling the plt.plot function with the specific parameters for each data.

```python
stock_a = [100,102,99,101,103,100,102]
stock_b = [90,95,102,104,105,103,109]
stock_c = [110,115,100,105,100,98,95]

years = [2016 + x for x in range (7)]
plt.title("Company Stocks")

plt.plot(years, stock_a, label="C1")
plt.plot(years, stock_b, label="C2")
plt.plot(years, stock_c, label="C3")

plt.xlabel("Year")
plt.ylabel("Stock Prices in USD")
# These labels are important to identify the lines for the stocks and can be displayed with plt.legend()
# We don't need specific x values even though we provided them with similar year values, it would have been fine otherwise too.
plt.legend(loc="lower right")
# The loc parameter isn't necessary but it can be used for custom location of legend as in this example. By default, location is set to upper right
plt.show()
```

# Subplots


```python
# Subplots are used when we want to have multiple charts in the same window and not individuals as above

style.use("default")
# Define x axis values with arange
x = np.arange(100)

# plt.figure(figsize=(20,20))
# This function is used to increase plot size and prevent overlap, but it doesn't work in Colab due to it's different rendering options. We can use plt.tight_layout() as in bottom.

# We will first define the grid of the plot, consider like a matrix, and each element represents a subplot in the grid.
fig, axs = plt.subplots(2,2)
# 2 rows, 2 columns; fig, axs means we have 1 figure with 4 axes i.e, [0,0],[0,1],[1,0],[1,1]

#axs is used like indexes in a matrix for each element and will be used to make subplots

axs[0,0].plot(x, np.sin(x))
axs[0,0].set_title("Sine Wave")
# axs[0,0] will be used to access a plot and set properties.
# we can set xlabels and ylabels as in other plots but in the format set_xlabels and set_ylabels, same goes for other properties

axs[0,1].plot(x, np.cos(x))
axs[0,1].set_title("Cosine Wave")

axs[1,0].plot(x, np.random.random(100))
axs[1,0].set_title("Random Numbers")

axs[1,1].plot(x, np.log(x))
axs[1,1].set_title("Log Wave")

plt.tight_layout()
# prevents overlapping of subplots and labels

fig.suptitle("Four Plots")
# sets a super title for the figure
plt.show()

plt.savefig("fourplots.png", dpi=500)
# saves the plot as a png image
# for some reason the downloaded image is just a blank screen, might be due to colab environment, will check later
```

# Surface Plot

A surface plot is a type of 3D plot that displays a three-dimensional surface where the height of the surface corresponds to the values of a function or dataset. It's used to visualize relationships and variations in data across two input dimensions and an output dimension.

A real-life example of a surface plot could be visualizing temperature variations over a geographical area. Imagine you have temperature measurements recorded at different coordinates (latitude and longitude) on the Earth's surface. You can create a surface plot to represent how temperatures change across the landscape.

```python
ax = plt.axes(projection="3d")

x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)

X,Y = np.meshgrid(x,y)

Z = np.sin(X) * np.cos(Y)

ax.plot_surface(X,Y,Z, cmap="Spectral")
# cmap is just used to define a color map
ax.set_title("3D Plot")
ax.set_xlabel("test")

plt.show()

```

# Animating Plots


```python
import random
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
heads_tails = [0,0]

for _ in range(100000):
  heads_tails[random.randint(0,1)] += 1
  plt.bar(["Heads", "Tails"], heads_tails, color=["red","blue"])
  plt.pause(0.001)
plt.show()

# it shows an animated plot with dynamic values of each coin flip, with outcome as either heads or tails.
# It doesn't work as an animation in colab however, it just shows individual plots for each iteration.
# This means, 100000 plots, so we won't test it here.

'''
heads_tails = [0, 0]: Initializes a list to keep track of the count of heads (index 0) and tails (index 1) obtained from the coin flips.

for _ in range(100000):: Iterates through a loop 100,000 times to simulate coin flips.

heads_tails[random.randint(0, 1)] += 1: Simulates a coin flip using random.randint(0, 1). It generates a random integer (0 or 1) to represent heads or tails and increments the corresponding count in the heads_tails list.

plt.bar(["Heads", "Tails"], heads_tails, color=["red", "blue"]): Creates a bar chart using Matplotlib's plt.bar() function. The x-axis has two categories: "Heads" and "Tails," and the heights of the bars represent the count of heads and tails. The color argument specifies that the "Heads" bar will be red and the "Tails" bar will be blue.

plt.pause(0.001): Pauses the plot briefly (0.001 seconds) to allow for dynamic updating. This creates an animation effect as the bars change with each iteration.

plt.show(): Displays the final bar chart after all iterations are complete.
'''
```
